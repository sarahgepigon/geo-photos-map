```
# Seattle - Space Needle
space_needle_lat = 47.620422
space_needle_lon = -122.349358
file_seattle = 'https://upload.wikimedia.org/wikipedia/commons/thumb/2/23/Space_Needle_2011-07-04.jpg/960px-Space_Needle_2011-07-04.jpg'
caption_seattle = 'Space Needle'

# New York City - Empire State Building
empire_state_lat = 40.748817
empire_state_lon = -73.985428
file_nyc = 'https://upload.wikimedia.org/wikipedia/commons/thumb/9/91/View_of_Empire_State_Building_from_Rockefeller_Center_New_York_City_dllu_Cropped.jpg/800px-View_of_Empire_State_Building_from_Rockefeller_Center_New_York_City_dllu_Cropped.jpg'
caption_nyc = 'Empire State Building'

# Bangkok - Wat Arun
wat_arun_lat = 13.743500
wat_arun_lon = 100.489000
file_bangkok = 'https://upload.wikimedia.org/wikipedia/commons/thumb/2/2a/%E0%B9%80%E0%B8%88%E0%B8%94%E0%B8%B5%E0%B8%A2%E0%B9%8C%E0%B8%9B%E0%B8%A3%E0%B8%B0%E0%B8%98%E0%B8%B2%E0%B8%99%E0%B8%97%E0%B8%A3%E0%B8%87%E0%B8%9B%E0%B8%A3%E0%B8%B2%E0%B8%87%E0%B8%84%E0%B9%8C%E0%B8%A7%E0%B8%B1%E0%B8%94%E0%B8%AD%E0%B8%A3%E0%B8%B8%E0%B8%932.jpg/2560px-%E0%B9%80%E0%B8%88%E0%B8%94%E0%B8%B5%E0%B8%A2%E0%B9%8C%E0%B8%9B%E0%B8%A3%E0%B8%B0%E0%B8%98%E0%B8%B2%E0%B8%99%E0%B8%97%E0%B8%A3%E0%B8%87%E0%B8%9B%E0%B8%A3%E0%B8%B2%E0%B8%87%E0%B8%84%E0%B9%8C%E0%B8%A7%E0%B8%B1%E0%B8%94%E0%B8%AD%E0%B8%A3%E0%B8%B8%E0%B8%932.jpg'
caption_bangkok = 'Wat Arun'
```

```
import folium, base64
from branca.element import Template, MacroElement

# ------------------------------------------------------------------
# 1.  Minimal data â€“ one row per photo / lat-lon pair
# ------------------------------------------------------------------
data = [
    {"lat": empire_state_lat, "lon": empire_state_lon, "file": file_nyc, "caption": caption_nyc},
    {"lat": space_needle_lat, "lon": space_needle_lon, "file": file_seattle, "caption": caption_seattle},
    {"lat": wat_arun_lat, "lon": wat_arun_lon, "file": file_bangkok, "caption": caption_bangkok}
]
m = folium.Map(location=[empire_state_lat, empire_state_lon], zoom_start=12,
#  tiles="https://{s}.basemaps.cartocdn.com/light_nolabels/{z}/{x}/{y}.png",
tiles ="cartodb positron",
    attr='&copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors &copy; <a href="https://carto.com/attributions">CARTO</a>'
)

for row in data:
    img_html = f"""
<div style='text-align:center;'>
  <img src="{row['file']}" style="max-width:50vw;max-height:40vh;border-radius:6px;">
  <p style="margin:4px 0 0;font-weight:600;">{row['caption']}</p>
</div>
"""


    # OPTION A â€“ show the *image* as a Tooltip (hover only, no click)
    tooltip = folium.Tooltip(img_html, sticky=False)

    # OPTION B â€“ same HTML wrapped in a normal Popup (for mobile taps)
    popup  = folium.Popup(img_html, max_width="auto")

    marker = folium.Marker(
        location=[row["lat"], row["lon"]],
        tooltip=tooltip,
        popup=popup,
        icon=folium.Icon(color="orange", icon="camera", prefix="fa")
        ).add_to(m)

m
```
